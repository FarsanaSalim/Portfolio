```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from genetic_selection import GeneticSelectionCV
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv('parkinsons.data')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()

df['name'] = Le.fit_transform(df['name'])
df.head()
```

```python
x = df.loc[:,df.columns != 'status']
y = df['status']
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.2,
                                                    stratify=y,
                                                    random_state=42)
scv = StratifiedKFold(n_splits=5)
selector = GeneticSelectionCV(RandomForestClassifier(n_estimators=100),
                              # svm.SVC(),
                              # XGBClassifier(),
                              # tree.DecisionTreeClassifier(criterion = 'gini'),
                              # GaussianNB(),
                              # LogisticRegression(C=10, max_iter=10000, random_state=42),
                              cv=scv,
                              verbose=2,
                              scoring="f1_macro",
                              max_features=5,
                              n_population=50,
                              crossover_proba=0.5,
                              mutation_proba=0.2,
                              n_generations=40,
                              crossover_independent_proba=0.5,
                              mutation_independent_proba=0.05,
                              tournament_size=3,
                              n_gen_no_change=10,
                              caching=True,
                              n_jobs=-1)
selector = selector.fit(X_train, y_train)
# get the selected features
cols = x.columns.tolist()
selected_feats = [cols[i] for i in np.where(selector.support_)[0]]
# train and test
clf = LogisticRegression(C=10, max_iter=10000, random_state=42)
clf.fit(X_train[selected_feats], y_train)
preds = clf.predict(X_test[selected_feats])
print("Model accuracy(in %):", metrics.accuracy_score(y_test, preds)*100)
print(classification_report(y_test, preds))
```

```python
###### Removing Constant Features using VarianceThreshold

from sklearn.feature_selection import VarianceThreshold
constant_filter = VarianceThreshold(threshold=0)

#Fit and transforming on train data
data_constant = constant_filter.fit_transform(df)
print(data_constant.shape)

#Extracting all constant columns using get support function of our filter
constant_columns = [column for column in df.columns
                    if column not in df.columns[constant_filter.get_support()]]

#No. of constant columns
print(len(constant_columns))

#Constant columns names:
for column in constant_columns:
    print(column)
```

```python
########### Removing Quasi-Constant Features

qcons_filter = VarianceThreshold(threshold=0.0001)

#Fit and transforming on train data
data_qcons = qcons_filter.fit_transform(df)
print(data_qcons.shape)

#Extracting all Quasi constant columns using get support function of our filter
qcons_columns = [column for column in df.columns
                    if column not in df.columns[qcons_filter.get_support()]]

#No. of Quasi constant columns
print(len(qcons_columns))

#Quasi Constant columns names:
for column in qcons_columns:
    print(column)
```

```python
data_qcons = df.drop(qcons_columns,axis=1)
data_qcons.shape
```

```python
x1 = data_qcons.loc[:,data_qcons.columns != 'status']
y1 = data_qcons['status']
```

```python
x1
```

```python
y1
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(x1,
                                                    y1,
                                                    test_size=1/4,
                                                    # stratify=y,
                                                    random_state=0)
```

```python
scv = StratifiedKFold(n_splits=5)
selector = GeneticSelectionCV(
                              # RandomForestClassifier(n_estimators=100),
                              # svm.SVC(),
                              XGBClassifier(),
                              # tree.DecisionTreeClassifier(criterion = 'entropy'),
                              # GaussianNB(),
                              # LogisticRegression(C=10, max_iter=10000, random_state=42),
                              cv=scv,
                              verbose=2,
                              scoring="f1_macro",
                              max_features=5,
                              n_population=50,
                              crossover_proba=0.5,
                              mutation_proba=0.2,
                              n_generations=40,
                              crossover_independent_proba=0.5,
                              mutation_independent_proba=0.05,
                              tournament_size=3,
                              n_gen_no_change=10,
                              caching=True,
                              n_jobs=-1)
selector = selector.fit(X_train1, y_train1)
# get the selected features
cols = x1.columns.tolist()
selected_feats = [cols[i] for i in np.where(selector.support_)[0]]
# train and test
clf = LogisticRegression(C=10, max_iter=10000, random_state=42)
clf.fit(X_train1[selected_feats], y_train1)
preds = clf.predict(X_test1[selected_feats])
print("Model accuracy(in %) after filtering features:", metrics.accuracy_score(y_test1, preds)*100)
print(classification_report(y_test1, preds))
```

## **VOTING**

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
# fit model no training data
model1 = XGBClassifier()
model1.fit(X_train1, y_train1)

# make predictions for test data
y_pred_xgboost1 = model1.predict(X_test1)
predictions1 = [round(value) for value in y_pred_xgboost1]
# evaluate predictions
accuracy = accuracy_score(y_test1, predictions1)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
# import voting classifier
from sklearn.ensemble import VotingClassifier

# create a voting classifier with hard voting
voting_classifier_hard = VotingClassifier(
    estimators = [('dtc', tree.DecisionTreeClassifier(random_state=42)),
                  ('rf', RandomForestClassifier()),
                  ('xgb', XGBClassifier())],
    voting='hard')

# create a voting classifier with soft voting
voting_classifier_soft = VotingClassifier(
    estimators = [('dtc', tree.DecisionTreeClassifier(random_state=42)),
                  ('rf', RandomForestClassifier()),
                  ('xgb', XGBClassifier())],
    voting='soft')

# make predictions with the hard voting model
voting_classifier_hard.fit(X_train1, y_train1)
y_pred_vch1 = voting_classifier_hard.predict(X_test1)

# make predictions with the soft voting model
voting_classifier_soft.fit(X_train1, y_train1)
y_pred_vcs1 = voting_classifier_soft.predict(X_test1)

# evaluate both models with the f-1 score
acc_vch = metrics.accuracy_score(y_test1, y_pred_vch1)
acc_vcs = metrics.accuracy_score(y_test1, y_pred_vcs1)

# print the accuracy scores
print('Accuracy of the hard voting classifier: {}'.format(np.round(acc_vch,4)))
print('Accuracy of the soft voting classifier: {}'.format(np.round(acc_vcs,4)))
```

### **BAGGING**

```python
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
```

```python
seed = 8
kfold = model_selection.KFold(n_splits = 5)

# initialize the base classifier
base_cls = tree.DecisionTreeClassifier()
# base_cls = RandomForestClassifier()
# base_cls = XGBClassifier()

# no. of base classifier
num_trees = 500

# bagging classifier
model = BaggingClassifier(base_estimator = base_cls,
                          n_estimators = num_trees,
                          random_state = seed)

results = model_selection.cross_val_score(model, X_train1, y_train1, cv = kfold)
for i in range(len(results)):
  print("Model " + str(i) + " Accuracy is " + str(results[i]))

print("Mean accuracy: " + str(results.mean()))
```
